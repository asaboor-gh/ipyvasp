# Quickstart

```python
import numpy as np
import matplotlib.pyplot as plt

import ipyvasp as ipv
```

Let's create some dummy POSCAR.

```python
poscar = ipv.POSCAR.new(
    np.eye(3),
    {"H": [[0.5, 0.5, 0.5],[0.0, 0.0, 0.0]]},
    scale = 5
)

ax = poscar.splot_lattice(fill = False, colors=['blue'],label='original')

poscar.transform(lambda a,b,c: (a+b, a-b, c)).splot_lattice(
    ax=ax, fill = False, colors=['red'], color='red', label='transformed',
)
plt.legend(ncol=2,frameon=False)
ax.view_init(azim=15, elev=30)
```

```python
poscar.splot_bz(fill = False).set_axis_off()
ax = poscar.splot_kpath([(0,0,0), (0.5,0,0), (0.5,0.5,0), (0,0,0), (0.5,0.5,0.5)],
    labels = ['', 'X', 'M', '', 'R'])
ax.view_init(azim=55, elev=30)
```

```python
fig = poscar.iplot_bz(color='blue')
ipv.iplot2html(fig)
```

We can interplate kpath based on current POSCAR, it gives correct distance between points.

```python
poscar.get_kpath([
    (0,0,0,''), 
    (0.5,0,0,'X'), 
    (0.5,0.5,0,'M'), 
    (0,0,0,''), 
    (0.5,0.5,0.5,'R')
    ], n=8)
```

```python
poscar.get_kmesh(2,2,2, weight=0)
```
